```python
n = int(input("Nhập số nguyên dương: "))

# kiểm tra n > 0
while n <= 0:
    n = int(input("Nhập lại (phải > 0): "))

# kiểm tra số nguyên tố
if n < 2:
    print(n, "không phải số nguyên tố.")
else:
    is_prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print(n, "là số nguyên tố.")
    else:
        print(n, "không phải số nguyên tố.")
```

```python
def reverse_number(n):
    r = 0
    while n > 0:
        r = r * 10 + (n % 10)
        n //= 10
    return r
def is_palindrome(n):
    return n == reverse_number(n)
n = int(input("Nhập n: "))

steps = 0
while not is_palindrome(n):
    n = n + reverse_number(n)
    steps += 1

print(steps, n)

```

```python
def all_distinct_digits(x):
    seen = set()
    while x > 0:
        d = x % 10
        if d in seen:
            return False
        seen.add(d)
        x //= 10
    return True
n = int(input("Nhập n: "))

found = False
x = 1

while x * x <= n:
    sq = x * x
    if all_distinct_digits(sq):
        print(sq)
        found = True
    x += 1

if not found:
    print("no number")
```

```python
def collatz_len(x):
    cnt = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3*x + 1
        cnt += 1
    return cnt
n = int(input("Nhập n: "))

best_x = 1
best_len = 1

for i in range(1, n + 1):
    L = collatz_len(i)
    if L > best_len or (L == best_len and i < best_x):
        best_len = L
        best_x = i

print(best_x, best_len)
```

```python
def sum_proper_divisors(n):
    s = 1  # 1 luôn là ước riêng của mọi số > 1
    i = 2
    while i * i <= n:
        if n % i == 0:
            s += i
            if n // i != i:
                s += (n // i)
        i += 1
    return s
a = int(input())
b = int(input())

if sum_proper_divisors(a) == b and sum_proper_divisors(b) == a:
    print("true")
else:
    print("false")
```
