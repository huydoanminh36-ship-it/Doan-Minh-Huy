```python
# Bài 3
import math

a, b, c = map(float, input("Nhap a b c: ").split())

# Kiểm tra có tạo thành cạnh tam giác không
if a + b <= c or a + c <= b or b + c <= a:
    print("Khong la ba canh cua mot tam giac")
else:
    # Phân loại tam giác
    if a == b == c:
        print("Tam giac deu")
    elif a == b or a == c or b == c:
        print("Tam giac can")
    elif a*a == b*b + c*c or b*b == a*a + c*c or c*c == a*a + b*b:
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")

    # Tính chu vi
    P = a + b + c

    # Tính diện tích theo Heron
    p = P / 2
    S = math.sqrt(p * (p - a) * (p - b) * (p - c))

    print("Chu vi =", P)
    print("Dien tich =", S)
```

```python
# Bài 5
import math

a, b, c = map(float, input("Nhap a b c: ").split())

if a == 0:
    # trở về phương trình bậc 1
    if b == 0:
        print("Vo nghiem" if c != 0 else "Vo so nghiem")
    else:
        print("Mot nghiem:", -c / b)
else:
    delta = b*b - 4*a*c
    if delta < 0:
        print("Vo nghiem")
    elif delta == 0:
        print("Nghiem kep:", -b / (2*a))
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print("Hai nghiem:", x1, x2)
```

```python
# Bài 8
a, b, m = map(float, input("Nhap a b m: ").split())
c, d, n = map(float, input("Nhap c d n: ").split())

D = a*d - b*c
Dx = m*d - b*n
Dy = a*n - m*c

if D == 0:
    if Dx == 0 and Dy == 0:
        print("He vo so nghiem")
    else:
        print("He vo nghiem")
else:
    x = Dx / D
    y = Dy / D
    print("x =", x)
    print("y =", y)
```

```python
# Bài 10
a, b = map(float, input("Nhap toa do tam duong tron (a b): ").split())
R = float(input("Nhap ban kinh R: "))

x, y = map(float, input("Nhap diem A(x y): ").split())

kc2 = (x - a)**2 + (y - b)**2   
R2 = R * R

if kc2 < R2:
    print("Diem nam trong duong tron")
elif kc2 == R2:
    print("Diem nam tren duong tron")
else:
    print("Diem nam ngoai duong tron")
```
